```python
# for loop 돌며 잘못된 row 찾기
# <> 채우기
for row in data_df.iterrows():
  total_sum = sum(row[2:])
  if total_sum != 1:
    print(row)
```

```python
# drop 이용해 삭제!
for row in data_df.itertuples():
  total_sum = sum(row[2:])
  if total_sum != 1:
    index = row[0]
    # <> 채우기
    data_df.drop(index, axis=0, inplace=True)
```

##### 특수 문자 제거하기

###### 특수 문자 체크

```python
import re

special_character_list = list()

# 숫자, 한글, 공백, !?,. 제외하고 어떤 문자들이 있는지 확인
# <> 채우기
special = re.compile("[^\s\!\?\.\,ㄱ-ㅎ가-힣0-9]")
special_list = list()

for row in data_df.itertuples():
  text = row[1]
  special_list.extend(re.findall(special, text))

set(special_list)
```

###### 특수문자 제거
- pandas.DataFrame.replace

```python
# replace 활용해 치환!
# <> 채우기
data_df.replace(regex=r'[^\s\!\?\.\,ㄱ-ㅎ가-힣0-9]', value='', inplace=True)
```

```python
# 확인
data_df
```

##### 앞, 뒤, 중복 공백 문자 처리하기

###### 공백문자 체크

```python
# 중복 공백, 앞, 뒤 공백 체크
# <> 채우기
double_space = re.compile('  +')
space_head = re.compile('^\s.+')
space_tail = re.compile('.+\s$')

# 공백 문제있는 utt 반환하는 함수 정의
def find_spaces(dataframe):
  for row in dataframe.itertuples():
    text = row[1]
    if re.findall(double_space, text):
      print(f"중복 공백 : {text}")
    if re.findall(space_head, text):
      print(f"앞 공백 : {text}")
    if re.findall(space_tail, text):
      print(f"뒷 공백 : {text}")

find_spaces(data_df)
```

###### 중복 공백 제거

```python
# replace, regex 활용해 중복 공백 -> 공백 하나로 치환
# <> 채우기
data_df.replace(regex="  +", value=" ", inplace=True)
```

```python
# 중복 공백 제거되었는지 확인
find_spaces(data_df)
```

###### 앞, 뒤 공백 삭제

```python
# 앞, 뒤 공백 삭제하기
# <> 채우기
# '' 안에 발화 컬럼 title을 입력해주세요. ex) utt
data_df['utt'] = data_df['utt'].str.strip()
```

```python
# 공백 문제 해결되었는지 확인
find_spaces(data_df)
```

###### 글자수 체크

```python
# 글자수 체크하는 함수 정의
def count_character(dataframe):
  for row in dataframe.itertuples():
    text = row[1]
    text = re.sub(' ', '', text)
    print(f"글자 수 : {len(text)}")
```

```python
count_character(data_df)
```

###### 글자수 필터링
- 각자 적정 글자수 정하기
- 참고 : [pandas.Series.apply](https://pandas.pydata.org/docs/reference/api/pandas.Series.apply.html)

```python
# min = 최소 글자수, max = 최대 글자수
# <> 채우기
min = 2
max = 25
data_df = data_df[data_df["utt"].apply(lambda x: len(x) >= min) & data_df["utt"].apply(lambda x: len(x) <= max)]
data_df.reset_index(drop=True, inplace=True)
```

```python
# 확인
data_df
```
