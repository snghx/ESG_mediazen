##### 앞, 뒤, 중복 공백 문자 처리하기

###### 공백문자 체크

```python
print(new_df["utt"])
```

```python
double_space = re.compile('  +')
space_head = re.compile('^\s.+')
space_tail = re.compile('.+\s$')

def find_spaces(dataframe):
  for row in dataframe.itertuples():
    text = row[1] #utt
    if re.findall(double_space, text):
      print(f"중복 공백 : {text}")
    if re.findall(space_head, text):
      print(f"앞 공백 : {text}")
    if re.findall(space_tail, text):
      print(f"뒷 공백 : {text}")

find_spaces(new_df)
```

###### 중복 공백 제거

```python
new_df.replace(regex="\s+", value=" ", inplace=True) #regex = ""로도 정규표현식 사용 가능
```

```python
find_spaces(new_df)
```

###### 글자수 체크
공백을 제거하고 체크하는 게 좋음

```python
def count_character(dataframe):
  for row in dataframe.itertuples():
    text = row[1]
    text = re.sub(' ', '', text) # 공백 제거 후
    print(f"글자 수 : {len(text)}")
```

```python
count_character(new_df)
```

###### 글자수 필터링
- 적정 글자수 최소 2글자 최대 20글자
- 참고 : [pandas.Series.apply](https://pandas.pydata.org/docs/reference/api/pandas.Series.apply.html)

```python
new_df = new_df[new_df["utt"].apply(lambda x: len(x) >= 2) & new_df["utt"].apply(lambda x: len(x) <= 20)]
new_df.reset_index(drop=True, inplace=True)

#series.apply(function) : 해당 series에 대해 매개변수로 들어간 함수를 모두 실행시켜줌
#df[True, False, True,,,]이렇게 적어주면 True인 값만 df안에 들어감
```
