```python
import pandas as pd

file_path = "/content/drive/MyDrive/중정처1주차/df_text.csv"
data_df = pd.read_csv(file_path, index_col=0, header=0)
data_df
```

```python
df_train = data_df.sample(frac=0.8, random_state=1) #random_state는 랜덤한 값을 정해둠 : 몇번을 돌리더라도 같은 값 나옴, frac : 표본추출 비율을 의미
print(df_train)
print()
print(f"length of train df: {len(df_train)}")
```

```python
df_remain = data_df.drop(df_train.index) #표본추출하고 난 나머지(20%)를 df_remain에 저장
df_eval = df_remain.sample(frac=0.5, random_state=1)
df_test = df_remain.drop(df_eval.index)
print(df_eval)
print()
print(f"length of test df: {len(df_eval)}")
print()
print(df_test)
print()
print(f"length of test df: {len(df_test)}")
```

```python
train_path = "/content/drive/MyDrive/중정처1주차/processed_train.csv"
eval_path = "/content/drive/MyDrive/중정처1주차/processed_eval.csv"
test_path = "/content/drive/MyDrive/중정처1주차/processed_test.csv"
df_train.to_csv(train_path)
df_eval.to_csv(eval_path)
df_test.to_csv(test_path)
```
