# 실시간 기상청 날씨 구하기

```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('http://www.weather.go.kr/weather/observation/currentweather.jsp')
# print(response)
# print(response.content)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
print(soup)
```

```python
table = soup.find('table', {'id':'weather_table'})
print(table)
```

```python
data = []                                       # 비어있는 리스트 : 지역, 기온, 습도

for tr in table.find_all('tr'):
    tds = list(tr.find_all('td'))

    for td in tds:
        if td.find('a'):
            point = td.find('a').text           # 지역명
            temp = tds[5].text                  # 기온
            humidity = tds[9].text              # 습도

            data.append([point, temp, humidity])

print(data)
```

```python
print('[지역, 기온, 습도]')
for w in data:
    print(w)
```
