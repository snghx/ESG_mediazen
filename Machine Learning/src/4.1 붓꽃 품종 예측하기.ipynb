## 사이킷런을 이용하여 붓꽃(iris) 데이터 품종 예측하기

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# 붓꽃 데이터세트 로딩
iris = load_iris()
print(iris)
```

```python
# iris 데이터세트에서 data 추출
iris_data = iris.data
print(iris_data)
```

```python
# iris.target은 레이블값을 구해옴(0, 1, 2)
# iris.target_names은 붓꽃 품종명 구해옴 (setosa, versicolor, virginica)
# iris.feature_names은 붓꽃 데이터 속성을 구해옴 ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
iris_label = iris.target
target_names = iris.target_names
feature_names = iris.feature_names
print('iris label:', iris_label)
print('iris target:', target_names)
print('feature_names:', feature_names)
```

```python
# 붓꽃 데이터 세트를 자세히 보기 위해 DataFrame으로 변환 
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df['label'] = iris.target             # label 열 추가함
iris_df.head(5)                            # 앞에서 부터 5개의 데이터 구해옴
```

```python
# train data 와 test data로 분할(8:2 비율)
x_train, x_test, y_train, y_test = train_test_split(iris_data, iris_label, 
                                                    test_size=0.2, random_state=11)
print('x_train:', x_train)
print('y_train:', y_train)
```

```python
# DecisionTreeClassifier 객체 생성 
model = DecisionTreeClassifier(random_state=11)

# 나이브 베이즈(Naive Bayes) 분류기 객체 생성
# model = GaussianNB()
```

```python
# 학습 데이터 세트로 학습 수행 
model.fit(x_train, y_train)
```

```python
# 테스트 데이터 세트로 예측 수행
pred = model.predict(x_test)
```

```python
# 예측 정확도
from sklearn.metrics import accuracy_score
print('예측 정확도: {0:.4f}'.format(accuracy_score(y_test,pred)))
```
